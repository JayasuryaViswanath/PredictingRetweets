```python
import time
while data.shape[0]<100:
    time.sleep(10)
    tweets = api.search_tweets('DonorsChoose', max_id=str(min(data['tweet_id'])))
    tweet_id=[]
    for i in range(0, len(tweets)):
        tweet_id.append(tweets[i]._json['id'])

    tweet_created=[]
    for i in range(0, len(tweets)):
        tweet_created.append(tweets[i]._json['created_at'])
    
    followers_count=[]
    for i in range(0, len(tweets)):
        followers_count.append(tweets[i]._json['user']['followers_count'])

    retweet_count=[]
    for i in range(0, len(tweets)):
        retweet_count.append(tweets[i]._json['retweet_count'])
    
    data_1=pd.DataFrame()
    data_1['tweet_id']=tweet_id
    data_1['created_at']=tweet_created
    data_1['followers_count']=followers_count
    data_1['retweet_count']=retweet_count
    data=pd.concat([data,data_1], ignore_index=True)
    print("Collecting .....",data.shape[0],"/","100")
  
    
    
```

```python
data
```

```python
data=data[['followers_count','retweet_count']]
```

```python
def category(x):
    if x>0:
        return 1
    else:
        return 0

data['retweet_count']=data['retweet_count'].apply(lambda x:category(x))
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.3)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(train[['followers_count']], train[['retweet_count']])
```

```python
clf.predict(test[['followers_count']])
```

```python
test
```

```python
test['prediction']=clf.predict(test[['retweet_count']])
```

```python
test
```

```python
import sklearn
sklearn.metrics.confusion_matrix(test[['retweet_count']], clf.predict(test[['followers_count']]))
```

```python
sklearn.metrics.accuracy_score(test[['retweet_count']], clf.predict(test[['followers_count']]))
```
